```
// Question 4: Even Odd array.


int[] array = new int[5];  // Declaring an array with pre-defined size
array[0] = 25;
array[1]= 47;
array[2]= 42;
array[3]= 56;
array[4]= 32;     // Element inputs of the array

int[] evenArray = new int[5];
int[] oddArray = new int[5];
                                 // Declaring two arrays for saving 'even' and 'odd' outputs                                  


for (int i = 0; i < 5; i++)
{
    Console.WriteLine($"The element {i + 1} = {array[i]}");  // printng out array elements

    if (array[i] % 2 == 0)       // if the remainder is zero then element would be even
    {
        evenArray[i] = array[i]; // storing element to 'evenArray'
    }
    else                         // when if condition is false, the element must be 'odd'
    {
        oddArray[i] = array[i];
    }
}

Console.WriteLine("\n\n=== The Even Array ===\n\n");
for (int i = 0; i < evenArray.Length; i++)           // for loop here is used to print out 'evenArray'
{
    if (evenArray[i] != 0)                    // if is used to restrict the printing of "zero"
    {
        Console.Write($"\t{evenArray[i]}");
    }
}

Console.WriteLine("\n\n=== The Odd Array ===\n\n");

for (int i = 0; i < oddArray.Length; i++)        // for loop here is used to print out 'oddArray'
{
    if (oddArray[i] != 0)        // if is used to restrict the printing of "zero"
    {
        Console.Write($"\t{oddArray[i]}");
    }
}
```
